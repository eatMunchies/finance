```python
# imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from data_handler import DataHandler
```

```python
handler = DataHandler()
data = handler.get_from_csv("VOO_8_18_25_to_8_22_25")
data['return'] = (data['close'] - data['open']) / data['open']
```

```python
fig, ax = plt.subplots(figsize=(10, 6))
ax.plot(data['date'], data['return'] * 100, color='red')
# ax.plot(data['date'], data['volume'], color='skyblue')
ax.set_xlabel('Date')
ax.set_ylabel('Value')
plt.show()
```
